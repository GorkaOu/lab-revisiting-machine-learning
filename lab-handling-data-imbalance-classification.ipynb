```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import export_text
```

```python
pd.set_option('display.max_columns',None)
data = pd.read_csv('clean_df.csv')
display(data)
display(data.shape)
```

```python
#I drop Target D to
X = data.drop(columns=['TARGET_B', 'TARGET_D'])
y = data['TARGET_B']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='first')
X_train_cat_encoded = enc.fit_transform(X_train_cat).toarray()
X_test_cat_encoded = enc.transform(X_test_cat).toarray()
```

```python
X_train = pd.concat([pd.DataFrame(X_train_num_scaled), pd.DataFrame(X_train_cat_encoded)], axis=1)
X_test = pd.concat([pd.DataFrame(X_test_num_scaled), pd.DataFrame(X_test_cat_encoded)], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
# create a logistic regression model
model = LogisticRegression(max_iter= 1000)

# fit the model to the training data
model.fit(X_train, y_train)

# make predictions on the testing data
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
print("Accuracy on the test data: {:.2f}%".format(accuracy * 100))
```

# Managing imbalance in the dataset

```python
y_train = y_train.reset_index().drop(['index'], axis=1)
```

```python
trainset= pd.concat([X_train, y_train], axis=1)
trainset
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

```python
category_0.shape
```

```python
category_1.shape
```

## Upsampling (oversampling)

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
data_upsampled['TARGET_B'].value_counts()
```

```python
X_train_upsampled = data_upsampled.drop(['TARGET_B'], axis =1)
y_train_upsampled = data_upsampled['TARGET_B'] 
```

```python
# create a logistic regression model
model = LogisticRegression(max_iter= 2000)
# fit the model to the training data
model.fit(X_train_upsampled , y_train_upsampled)

# make predictions on the testing data
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
print("Accuracy on the test data: {:.2f}%".format(accuracy * 100))
```

## Downsampling (undersampling)

```python
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

```python
X_train_downsampled = data_downsampled.drop(['TARGET_B'], axis =1)
y_train_downsampled = data_downsampled['TARGET_B'] 
```

```python
# create a logistic regression model
model = LogisticRegression(max_iter= 1000)
# fit the model to the training data
model.fit(X_train_downsampled , y_train_downsampled)

# make predictions on the testing data
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
print("Accuracy on the test data: {:.2f}%".format(accuracy * 100))
```
