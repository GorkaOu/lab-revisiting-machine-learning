```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('learningSet.csv')
data.shape
```

```python
numerical = data.select_dtypes(np.number)
categorical = data.select_dtypes(object)
```

```python
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
categorical
```

```python
# I check categoricals and there is only a column with NaN values
nulls_percent_categorical = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_categorical.columns = ['column_name', 'nulls_percentage']
nulls_percent_categorical[nulls_percent_categorical['nulls_percentage']!=0]  
```

```python
# The drop list will be empty
columns_above_threshold = nulls_percent_categorical[nulls_percent_categorical['nulls_percentage']>0.85]
```

```python
# I add to the drop list the columns OSOURCE and ZIP
drop_columns_list = list(columns_above_threshold['column_name'])
columns = ['OSOURCE', 'ZIP']
for col in columns:
    drop_columns_list.append (col)
print(drop_columns_list)
```

```python
len (drop_columns_list)
```

```python
nullframe = pd.DataFrame(categorical.isna().sum(), columns = ['nulls'])
pd.set_option('display.max_rows', 100)
null_df = nullframe[nullframe['nulls'] != 0]
null_df
```

```python
# We are going to drop like in class
columns_2 = ['NOEXCH', 'MDMAUD', 'MDMAUD_R', 'MDMAUD_F','MDMAUD_A', 'PETS','PEPSTRFL']
for column in columns_2:
    drop_columns_list.append(column)
```

```python
# We will keep the column RFA_2X. We will delete rest of the columns 

for col_name in categorical.columns:
    if "RFA" in col_name:
        drop_columns_list.append(col_name) 
```

```python
drop_columns_list.remove('RFA_2R')
drop_columns_list.remove('RFA_2A')
```

```python
categorical = categorical.drop(drop_columns_list, axis=1)
```

```python
categorical['DOMAIN'].value_counts(dropna=False)
categorical['DOMAIN'] = categorical['DOMAIN'].fillna('R2').astype(str)
categorical['DOMAIN'].dtypes
```

```python
# I = Inferred from Date of Birth Field so we add there
categorical['AGEFLAG'].value_counts(dropna=False)
categorical['AGEFLAG'] = categorical['AGEFLAG'].fillna('I').astype(str)
```

```python
# We can maintain if we give 0 value to NaN values because it means No code has been assigned or did not match at any level.
categorical['GEOCODE'].value_counts(dropna=False)
categorical['GEOCODE'] = categorical['GEOCODE'].fillna('0').astype(str)
```

```python
categorical['LIFESRC'].value_counts(dropna=False)
# We add a new value call 0 which means we don´t check third-party data source
categorical['LIFESRC'] = categorical['LIFESRC'].fillna('0').astype(str)
```

```python
# I distinguish in gender between F,M and others 
categorical['GENDER'].value_counts().index
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F').astype(str)
```

```python
df = pd.DataFrame(categorical['GENDER'].value_counts()).reset_index()

df.columns = ['GENDER', 'count']
other_gender = list(df[df['count']<3000]['GENDER'])

#other_states
def clean_gender(x):
    if x in other_gender:
        return 'others'
    else:
        return x
    
categorical['GENDER'] = list(map(clean_gender, categorical['GENDER']))
```

```python
# # other way
# def clean_gender(x):
#     if x in ['F','M']
#         return x
#     else:
#         return 'others'
```

```python
categorical['GENDER'].value_counts()
```

```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
sorted(categorical['CLUSTER'].value_counts().index)
categorical['CLUSTER'] = categorical['CLUSTER'].fillna('40').astype(str)
```

```python
categorical['HOMEOWNR'].value_counts(dropna=False)
categorical['HOMEOWNR'] = categorical['HOMEOWNR'].fillna('U').astype(str)
```

```python
categorical['DATASRCE'].value_counts(dropna=False)
# We add a new value call 0 which means we don´t check third-party data source
categorical['DATASRCE'] = categorical['DATASRCE'].fillna('0').astype(str)
```

**Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.**

```python
# In my case, GEOCODE2 is not numerical, is categorical.
categorical['GEOCODE2'].value_counts(dropna=False)
```

```python
# I will add them to the category "A" because they don´t make the diference and "A" is the usual answer
categorical['GEOCODE2'] = categorical['GEOCODE2'].fillna('A')
categorical['GEOCODE2'] = categorical["GEOCODE2"].replace({" ": "A"})
categorical['GEOCODE2'].value_counts()
```

```python
# Indicates total number of known times the donor has responded to a mail order offer other than PVA's.
numerical['WEALTH1'].value_counts(dropna=False)
```

```python
# Usually, they don´t answer to the emails so we can assume that we can add to "0,0"
numerical['WEALTH1'] = numerical['WEALTH1'].fillna(0.0)
numerical['WEALTH1'].value_counts(dropna=False)
```

```python
# There are 132 NaN values in the next 3 columns. I try to find the meaning of this columns but I didn´t find anything so  
# I add NaN values to the first variable.
numerical['ADI'].value_counts(dropna=False)
```

```python
numerical['ADI'] = numerical['ADI'].fillna(13)
numerical['ADI'].value_counts(dropna=False)
```

```python
# There are 132 NaN values I will add to 803
numerical['DMA'].value_counts(dropna=False)
```

```python
numerical['DMA'] = numerical['DMA'].fillna(803)
numerical['DMA'].value_counts(dropna=False)
```

```python
# There are 132 NaN values I will add to 0.0
numerical['MSA'].value_counts(dropna=False)
```

```python
numerical['MSA'] = numerical['MSA'].fillna(0)
numerical['MSA'].value_counts(dropna=False)
```

```python
# Check again NaN values, there are 87 columns.
nullframe_num = pd.DataFrame(numerical.isna().sum(), columns = ['nulls'])
pd.set_option('display.max_rows', 100)
null_df = nullframe_num[nullframe_num['nulls'] != 0]
display (null_df, null_df.shape)
```

```python
# I will drop the columns with the next words. They have information about when the promotion was mailed, the girf was received,
# amount of the gift, number of children and the las 6 show the number of known times the donor has responded to other type of 
# mail order offers. For the last one, I maintain WEALTH1.

drop_list = []
col_to_drop = ["ADATE", "RDATE", 'RAMNT_', 'NUMCHLD','MBCOLECT','MBGARDEN','MBCRAFT','MAG','PUB','MBB']

# loop through each column in the dataframe
for col in numerical.columns:
    
    # check if any value in col_to_drop is a substring of the column name
    if any(val in col for val in col_to_drop):
        
        # if a match is found, append the column name to drop_list
        drop_list.append(col)
```

```python
drop_list
```

```python
numerical = numerical.drop(columns=drop_list)
```

```python
# Here I made same as in class.
numerical['AGE'] = numerical["AGE"].fillna(np.mean(numerical['AGE']))
numerical['INCOME'] = numerical['INCOME'].fillna(5.0)
numerical['CLUSTER2'] = numerical['CLUSTER2'].fillna(np.ceil(np.mean(numerical['CLUSTER2'])))
numerical['WEALTH2'] = numerical['WEALTH2'].astype('object')
numerical['WEALTH2'] = numerical['WEALTH2'].fillna('5.0')
numerical['TIMELAG'].median()
numerical['TIMELAG'] = numerical['TIMELAG'].fillna(6.0)
# In this case I decided to fill with 0.0 because I think there isn´t a second time for a gift
numerical['NEXTDATE'] = numerical['NEXTDATE'].fillna(0.0)
```

```python
# Change some of them to categorical
numerical['TARGET_B'] = numerical['TARGET_B'].astype('object')
numerical['WEALTH1'] = numerical['WEALTH1'].astype('object')
```

```python
# I don´t have NaN values now.
nullframe_num = pd.DataFrame(numerical.isna().sum(), columns = ['nulls'])
pd.set_option('display.max_rows', 100)
null_df = nullframe_num[nullframe_num['nulls'] != 0]
null_df
```
